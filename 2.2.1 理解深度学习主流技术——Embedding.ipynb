```python
import tensorflow as tf
import numpy as np
```

```python
model = tf.keras.Sequential()
# 添加Embedding层将1000维映射到64维
model.add(tf.keras.layers.Embedding(1000, 64, input_length=16))
# 编译模型
model.compile('adam', 'mse')
```

```python
# 初始化4个列表（4句话），每个列表有16个数字（单词），每个单词的索引范围在0-1000
input_array = np.random.randint(1000, size=(4, 16))
print(f'模型输入形状：{input_array.shape}')
print(f'模型输入：{input_array}\n')
# 预测模型
output_array = model.predict(input_array)
print(f'模型输出形状：{output_array.shape}') # (32, 10, 64)
print(f'模型输出：{output_array}') # (32, 10, 64)
```
